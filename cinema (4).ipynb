# Les titres des films sortient entre 2000 et 2005

```python
for x in db.top45.find({"année de sortie": {"$gte": 2000.0, "$lte": 2005.0 }}, {"_id":0, "titre":1}).sort("année de sortie", -1):
    pprint(x)
```

# Nombre des films par année

```python
project= { "$project" : {"_id" : 1, "totalfilm" : 1} }
group = {"$group" : {"_id" : "$année de sortie", "totalfilm" : {"$sum" : 1 } } }
sort = { "$sort":{"totalfilm":-1}}
for el in db.top45.aggregate([group, project, sort]):
    pprint(el)
```

# Export collection top45 into json file

```python
from bson.json_util import dumps

if __name__ == '__main__':
    client = MongoClient()
    db = client.culture
    cursor = top45.find({})
    with open('top45_collection.json', 'w') as file:
        file.write('[')
        for document in cursor:
            file.write(dumps(document))
            file.write(',')
        file.write(']')
```
